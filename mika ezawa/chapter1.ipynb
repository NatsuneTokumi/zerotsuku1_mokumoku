```python
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.plot(x,y1,label='sin')
plt.plot(x,y2,linestyle = '--', label='cos')
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin & cos')
plt.legend() #たぶん凡例
plt.show()
```

```python
def AND(x1, x2):
    w1, w2, theta = 0.5, 0.5, 0.7
    tmp = x1*w1 + x2*w2
    if tmp <= theta:
        return 0
    elif tmp > theta:
        return 1
```

```python
AND(0,1)
```

```python
AND(1,1)
```

```python
AND(1,0)
```

```python
AND(0,0)
```
